# Data Preprocessing

Handling Null Values

```python
df.isna().sum()
```

```python
df['weight'] = df['weight'].str.replace('lbs', '')

# Convert the weight column to numeric values
df['weight'] = pd.to_numeric(df['weight'])
```

```python
df.head()
```

```python
def convert_height(height):
    feet = int(height.split("'")[0])
    inches = int(height.split("'")[1].replace('"', ''))
    total_inches = feet * 12 + inches
    return total_inches

df['height'] = df['height'].apply(convert_height)
```

```python
df.head()
```

```python
num_with_na=['weight', 'ball_controlling_skills', 'jumping_skills', 'penalties_conversion_rate', 'mental_strength',
            'shot_accuracy','behaviour_rating', 'matches_played', 'fitness_rating','years_of_experience' ]
for i in num_with_na:
    df[i]=df[i].fillna(df[i].median())
```

```python
cat_with_na = ['gender','strong_foot','coaching']
for i in cat_with_na:
    df[i]=df[i].fillna(df[i].mode()[0])
```

```python
from sklearn.preprocessing import MinMaxScaler
# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Fit and transform the scaler on the dataset
scaled = scaler.fit_transform(df)

# Create a new DataFrame with scaled values
df = pd.DataFrame(scaled, columns=df.columns)
```

```python
df.head()
```

```python
x=df.drop('selection',axis=1)
y=df['selection']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(xtrain,ytrain)
ypredlr=lr.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
print("Accuracy=",accuracy_score(ytest,ypredlr))
print("Precision=",precision_score(ytest,ypredlr))
print("recall_score=",recall_score(ytest,ypredlr))
print("f1_score=",f1_score(ytest,ypredlr))
```

```python
confusion_matrix(ytest,ypredlr)
```

```python
from sklearn.neighbors import KNeighborsClassifier
metric = []
neighbors = np.arange(2,15)

for k in neighbors:
    classifier = KNeighborsClassifier(n_neighbors = k, metric = 'minkowski',p=2)
    classifier.fit(xtrain,ytrain)
    y_pred_knn = classifier.predict(xtest)
    acc = accuracy_score(ytest,y_pred_knn)
    metric.append(acc)
```

```python
plt.plot(neighbors,metric,"o-")
plt.xlabel("kvalue")
plt.xlabel("accuracy")
plt.grid()
```

```python
classifier = KNeighborsClassifier(n_neighbors = 11, metric = 'minkowski',p=2)
classifier.fit(xtrain,ytrain)
y_pred_knn = classifier.predict(xtest)
```
